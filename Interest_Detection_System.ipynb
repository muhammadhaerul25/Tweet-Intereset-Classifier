###**Remove Punctuations**

```python
string.punctuation
```

```python
all = all.rename(columns={"Text": "Tweet"})
```

```python
all.dtypes
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

all['Tweet_punct'] = all['Tweet'].apply(lambda x: remove_punct(x))
all.head(10)
```

###**Tokenization**

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

all['Tweet_tokenized'] = all['Tweet_punct'].apply(lambda x: tokenization(x.lower()))
all.head()
```

###**Remove stopwords**

```python
import nltk
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
stopword = stopwords.words('english')
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
    
all['Tweet_stopwords'] = all['Tweet_tokenized'].apply(lambda x: remove_stopwords(x))
all.head(10)
```

```python
all.head(5)
```

##**Stemming and Lammitization**

```python
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

all['Tweet_stemmed'] = all['Tweet_stopwords'].apply(lambda x: stemming(x))
all.head()
```

```python
import nltk
nltk.download('omw-1.4')
```

```python
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

all['Tweet_lemmatized'] = all['Tweet_stopwords'].apply(lambda x: lemmatizer(x))
all.head()
```

```python
#data cleaning function (include all of them)
def clean_text(text):
    text_lc = "".join([word.lower() for word in text if word not in string.punctuation]) # remove puntuation
    text_rc = re.sub('[0-9]+', '', text_lc)
    tokens = re.split('\W+', text_rc)    # tokenization
    text = [ps.stem(word) for word in tokens if word not in stopword]  # remove stopwords and stemming
    return text
```

###**Vectorisation**

```python
countVectorizer = CountVectorizer(analyzer=clean_text) 
countVector = countVectorizer.fit_transform(all['Tweet'])
print('{} Number of tweets has {} words'.format(countVector.shape[0], countVector.shape[1]))
print(countVectorizer.get_feature_names())
```
